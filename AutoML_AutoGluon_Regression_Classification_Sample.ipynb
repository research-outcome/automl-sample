### Regression

#### Load Data

```python
#Define the label column
label = 'price_usd'
```

```python
#Load the initial unprocessed data
cars = pd.read_csv('cars.csv')
```

```python
cars.info()
```

```python
# Split the data to test and train
from sklearn.model_selection import train_test_split
cars_train, cars_test = train_test_split(cars, test_size=0.2, random_state=42)
cars_test_labels = cars_test[label]
cars_test_nolabel = cars_test.drop([label], axis=1)
cars_train_labels = cars_train[label]
cars_train_nolabel = cars_train.drop([label], axis=1)
cars_train.head()
```

```python
# Import AutoGluon Libraries
from autogluon.tabular import TabularDataset, TabularPredictor
# Create an instance and then fit the train data
regression = TabularPredictor(label=label, path='.').fit(cars_train)
```

```python
# Check the performance using test data
regression_pred = regression.predict(cars_test_nolabel)
regression_perf = regression.evaluate_predictions(y_true=cars_test_labels, y_pred=regression_pred, auxiliary_metrics=True)
```

```python
#Show the leaderboard
regression.leaderboard(cars_test, silent=True)
```

```python
#Show the best model
regression.get_model_best()
```

### Classification

```python
#Define the label column
label = 'Survived'
```

```python
#Load the initial unprocessed data
titanic = pd.read_csv('titanic.csv')
```

```python
# Check the performance using test data
classification_pred = classification.predict(titanic_test_nolabel)
classification_perf = classification.evaluate_predictions(y_true=titanic_test_labels, y_pred=classification_pred, auxiliary_metrics=True)
```
